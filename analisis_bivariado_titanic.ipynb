**Descripción:** En este notebook se analiza la relación entre la variable objetivo 'Survived' y otras cuatro variables del dataset Titanic para entender los factores que influyeron en la supervivencia.

## 1. Selección de Variables y Justificación
Las cuatro variables elegidas para el análisis bivariado con **'Survived'** son:
- **Age (Edad):** Puede influir en la probabilidad de supervivencia, ya que es posible que los niños hayan tenido prioridad en los rescates.
- **Fare (Tarifa):** Puede indicar diferencias socioeconómicas y el acceso a mejores zonas del barco.
- **Pclass (Clase del pasajero):** Refleja el estatus socioeconómico, que podría haber impactado la tasa de supervivencia.
- **Sex (Género):** Puede estar relacionado con la regla 'mujeres y niños primero'.

```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('titanic.csv')

# Mostrar las primeras filas
df.head()
```

## 2. Análisis Bivariado

```python
# Análisis de correlación entre variables numéricas y Survived
corr_matrix = df[['Survived', 'Age', 'Fare', 'Pclass']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Correlación entre Variables Numéricas')
plt.show()
```

## Interpretación del mapa de calor

El código genera un mapa de calor (heatmap) que muestra la correlación entre las variables 'Survived', 'Age', 'Fare' y 'Pclass' del dataset Titanic.  Cada celda del mapa representa la correlación entre dos variables.  

**Valores cercanos a 1:** indican una correlación positiva fuerte.  Es decir, cuando una variable aumenta, la otra también tiende a aumentar.

**Valores cercanos a -1:** indican una correlación negativa fuerte.  Cuando una variable aumenta, la otra tiende a disminuir.

**Valores cercanos a 0:** indican una correlación débil o inexistente.

## Correlación entre 'Survived' y otras variables:  

Como vemos en el mapa una correlación negativa entre 'Survived' y 'Pclass' de -0.34 sugiere que los pasajeros de clases más bajas tenían menor probabilidad de sobrevivir.  

Vemos una correlación positiva débil entre 'Survived' y 'Fare' de 0.26 podría indicar que los pasajeros que pagaron tarifas más altas tuvieron una mayor probabilidad de sobrevivir.

Vemos una correlación negativa debil entre 'Survived' y 'Age' de 0.08 podría mostrar si la edad influyó en las tasas de supervivencia (si es negativa, podría indicar que los más jóvenes tenían mayor probabilidad de supervivencia).

El mapa también mostrará cómo se relacionan entre sí las variables 'Age', 'Fare' y 'Pclass'.  Por ejemplo, podría haber una correlación negativa entre 'Pclass' y 'Fare', lo que indicaría que los pasajeros de clases más altas pagaron tarifas más elevadas.

El mapa de correlación proporciona una visualización rápida de las relaciones lineales entre las variables.  Es importante recordar que la correlación no implica causalidad.  Una fuerte correlación entre dos variables no significa necesariamente que una cause la otra.


```python
# Boxplot:  'Age' ' vs 'Survived'
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Survived'], y=df['Age'], palette='coolwarm')
plt.title('Distribución de la variable Age según la variable Survived')
plt.xlabel('Supervivencia (0 = No, 1 = Sí)')
plt.ylabel('Edad')
plt.show()
```

## Interpretación del gráfico de cajas Age vs Survived

El gráfico de caja (boxplot) "Distribución de la variable Age según la variable Survived" muestra la distribución de las edades de los pasajeros del Titanic, separando a los sobrevivientes (Survived = 1) de los que no sobrevivieron (Survived = 0).

**Mediana:** La línea horizontal dentro de cada caja representa la mediana de la edad para cada grupo (sobrevivientes y no sobrevivientes).  Si la mediana de los sobrevivientes es menor que la de los no sobrevivientes, indica que, en promedio, los sobrevivientes tienden a ser más jóvenes.

**Cuartiles:** Las líneas superior e inferior de cada caja indican el tercer cuartil (75%) y el primer cuartil (25%) de las edades, respectivamente. La extensión de la caja muestra la dispersión de la mitad de los datos.

**Bigotes:** Los "bigotes"  se extienden hasta 1.5 veces el rango intercuartílico (IQR) desde los cuartiles. Los puntos fuera de los bigotes son valores atípicos.

**Comparación entre grupos:** La comparación visual de las cajas nos permite identificar si existe una diferencia significativa en las distribuciones de edades entre sobrevivientes y no sobrevivientes.

El gráfico permite visualizar si la edad influyó en la probabilidad de supervivencia.
Si hay una diferencia notable en las medianas y la extensión de las cajas entre los dos grupos, sugiere que la edad fue un factor relevante.  
La presencia de valores atípicos también puede aportar información sobre casos específicos.

## Interpretación del gráfico de Bigotes Fare vs Survived

El gráfico de caja ("boxplot") "Distribución de Fare según Survived" muestra la distribución de las tarifas pagadas por los pasajeros, separando a los sobrevivientes (Survived = 1) de los que no lo lograron (Survived = 0).

**Mediana:** La línea horizontal dentro de cada caja representa la mediana de la tarifa para cada grupo.  Si la mediana de los sobrevivientes es mayor que la de los no sobrevivientes, indica que, en promedio, los sobrevivientes tendieron a pagar tarifas más altas.

**Cuartiles:** Las líneas superior e inferior de cada caja indican el tercer cuartil (75%) y el primer cuartil (25%) de las tarifas, respectivamente. La extensión de la caja muestra la dispersión de la mitad de los datos.  Una caja más grande indica una mayor variabilidad en las tarifas pagadas dentro de ese grupo.

**Bigotes:** Los "bigotes" se extienden hasta 1.5 veces el rango intercuartílico (IQR) desde los cuartiles. Los puntos fuera de los bigotes son valores atípicos (outliers), que representan tarifas inusualmente altas o bajas.

**Valores Atípicos:** La presencia de valores atípicos en uno u otro grupo puede ser relevante. En el caso de las tarifas, un valor atípico muy alto en el grupo de sobrevivientes podría indicar que pasajeros con boletos muy costosos tuvieron mayor probabilidad de supervivencia.

**Comparación entre grupos:** La comparación visual de las cajas permite identificar si existe una diferencia significativa en las distribuciones de tarifas entre sobrevivientes y no sobrevivientes.  Si las cajas están muy separadas, con medianas distintas y pocos valores atípicos en común, sugiere que la tarifa fue un factor importante para la supervivencia.

El gráfico ayuda a determinar si la tarifa del pasaje influyó en la probabilidad de supervivencia.  Si se observa una diferencia clara en la mediana y la distribución de las tarifas entre ambos grupos, se podría concluir que la tarifa fue un factor relevante.  La presencia y ubicación de los valores atípicos también proveen información valiosa para el análisis.


```python
# Gráfico de barras: Survived según Pclass
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='coolwarm')
plt.title('Survived según Pclass')
plt.xlabel('Pclass')
plt.ylabel('Passengers count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

## Interpretación del Gráfico El gráfico de barras "Survived según Pclass"

El gráfico de barras "*Survived según Pclass*" muestra la cantidad de pasajeros que sobrevivieron y no sobrevivieron al desastre del Titanic, agrupados por la clase del pasaje (Pclass).

**Eje X (Pclass):** Representa las tres clases de pasajeros (1, 2 y 3).
**Eje Y (Passengers count):** Muestra la cantidad de pasajeros en cada categoría.
**Colores (hue = Survived):** Las barras se dividen en dos colores, uno para los pasajeros que sobrevivieron (Yes) y otro para los que no sobrevivieron (No).

Al observar el gráfico, se puede apreciar la relación entre la clase del pasajero y su probabilidad de supervivencia.

***Se observa que:***
**Pclass 1:**  La clase 1 tiene una mayor cantidad de sobrevivientes que de no sobrevivientes.  Esto sugiere que los pasajeros de primera clase tuvieron una mayor probabilidad de sobrevivir.

**Pclass 2:**  En la clase 2, la cantidad de sobrevivientes y no sobrevivientes es más parecida que en la clase 1, aunque todavía se aprecia una mayor cantidad de sobrevivientes.

**Pclass 3:**  En la clase 3, la cantidad de no sobrevivientes es significativamente mayor que la de sobrevivientes.  Esto indica que los pasajeros de tercera clase tuvieron una menor probabilidad de supervivencia.

La **(Pclass) **parece haber sido un factor determinante en la supervivencia en el Titanic.  Los pasajeros de clases más altas (1 y 2) tuvieron una mayor probabilidad de supervivencia en comparación con los pasajeros de tercera clase. Esto concuerda con la correlación negativa observada en el mapa de calor entre 'Survived' y 'Pclass'.


```python
# Gráfico de barras: Survived según Sex
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Sex', hue='Survived', palette='coolwarm')
plt.title('Survived según Sex')
plt.xlabel('Sex')
plt.ylabel('Passengers count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

## Interpretación del Gráfico El gráfico de barras "Survived según Sex"

El gráfico de barras "Survived según Sex" muestra la cantidad de pasajeros que sobrevivieron y no sobrevivieron al desastre del Titanic, separados por sexo.

**Eje X (Sex):**  Muestra los dos géneros: masculino y femenino.
**Eje Y (Passengers count):** Indica la cantidad de pasajeros en cada categoría.
**Colores (hue = Survived):**  Las barras se dividen en dos colores, representando a los que sobrevivieron (Yes) y a los que no (No). La leyenda indica qué color corresponde a cada caso ('No' para fallecidos y 'Yes' para sobrevivientes).

Al observar el gráfico, podemos comparar la tasa de supervivencia entre hombres y mujeres:

**Mujeres:** Se aprecia una cantidad considerablemente mayor de mujeres sobrevivientes en comparación con las que no lo lograron.  Esto visualiza la aplicación de la regla "mujeres y niños primero" durante el proceso de evacuación.

**Hombres:** Se observa una mayor cantidad de hombres que no sobrevivieron. La diferencia entre sobrevivientes y no sobrevivientes es mucho más pronunciada en el caso de los hombres que en el de las mujeres.

El gráfico de barras evidencia claramente que el género fue un factor determinante en la supervivencia del desastre, favoreciendo a las mujeres.
